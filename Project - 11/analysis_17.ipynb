```python
# Rename the columns
df.rename(columns={'Height':'Height_cm'},inplace=True)
df.rename(columns={'Weight':'Weight_kg'},inplace=True)
```

```python
#Convert the height in foot into cm
df['Height_cm']=df['Height_cm'].str.replace("'",".")
df['Height_cm']=df['Height_cm'].astype('float')
df['Height_cm']=df['Height_cm']*30.48
df
```

```python
# Convert the Preferred Foot into numerical
df['Preferred Foot']=df['Preferred Foot'].map({'Right':1,'Left':0})
df
```

```python
df.drop(['Real Face'],axis=1,inplace=True)
df
```

```python
df.rename(columns={'Value':'Value_euro'},inplace=True)
df.rename(columns={'Wage':'Wage_euro'},inplace=True)
df
```

```python
# Remove the currency symbol
df['Value_euro']=df['Value_euro'].str.replace('€','')
df['Wage_euro']=df['Wage_euro'].str.replace('€','')
```

```python
# Function to convert the value into numerical 
def value_to_int(df_value):
    try:
        value = float(df_value[1:-1])
        suffix = df_value[-1:]

        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value
```

```python
df['Value_euro'] = df['Value_euro'].apply(value_to_int)
df['Wage_euro'] = df['Wage_euro'].apply(value_to_int)
df
```

```python
# Predicting the overall rating of the player using the Linear Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df_model=df.drop(['Name','Nationality','Club','Overall','Potential'],axis=1)
df_model
```

```python
# Splitting the dataset into training and testing set
X=df_model.drop(['Best Overall Rating','Jersey Number','Best Position'],axis=1)
y=df_model['Best Overall Rating']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X.isnull().sum()
```

```python
# Importing the model
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
# Predicting the model
y_pred=model.predict(X_test)
```

```python
# Checking the mean squared error
mean_squared_error(y_test,y_pred)
```

```python
# Checking the accuracy of the model
accuracy=model.score(X_test,y_test)
print(accuracy*100,'%')
```

```python
# Importing Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```
