```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = pd.read_csv('social media influencers - youtube.csv')
data
```

```python
data.info()
```

```python
data=data.drop(['Category', 'Audience Country', 'avg likes','avg comments'],axis=1)
data
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# Function to convert values
def convert_to_number(x):
    if x.endswith('M'):
        return float(x[:-1]) * 1_000_000
    elif x.endswith('K'):
        return float(x[:-1]) * 1_000
    else:
        return float(x)
```

```python
data['Subscribers'] = data['Subscribers'].apply(convert_to_number)
```

```python
data['avg views']=data['avg views'].apply(convert_to_number)
```
