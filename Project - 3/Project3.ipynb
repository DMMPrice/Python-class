```python
# Plot a graph with Guest Country and Guest Type
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white")
plt.figure(figsize=(30,20))
sns.countplot(x='Guest_Country',data=df,hue='Guest_Type')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot a graph with Guest Country and Booking Channel using plotly
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white")
plt.figure(figsize=(20,10))
sns.countplot(x='Guest_Country',data=df,hue='Booking_Channel')
plt.xticks(rotation=90)
plt.show()
```
